```python
# Question: Area chart for yearly profit.

import matplotlib.pyplot as plt

years = [2018,2019,2020,2021,2022,2023]
profit = [20,25,30,35,40,50]

plt.fill_between(years, profit, color='green', alpha=0.5)

plt.title("Profit Growth")
plt.xlabel("Year")
plt.ylabel("Profit")

plt.show()

# Output:
# Area chart showing growth trend.
```

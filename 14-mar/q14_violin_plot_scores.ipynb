```python
# Question: Violin plot for exam scores.

import matplotlib.pyplot as plt
import numpy as np

A = np.random.normal(70,10,100)
B = np.random.normal(65,12,100)
C = np.random.normal(75,8,100)

plt.violinplot([A,B,C])
plt.xticks([1,2,3],["Class A","Class B","Class C"])

plt.title("Score Distribution")
plt.xlabel("Classes")
plt.ylabel("Scores")

plt.show()

# Output:
# Violin plot showing score distribution.
```

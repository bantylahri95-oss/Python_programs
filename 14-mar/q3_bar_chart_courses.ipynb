```python
# Question: Create a Bar Chart for student enrollment.

import matplotlib.pyplot as plt

courses = ["Python","Java","Data Science","Web Dev"]
students = [120,90,100,80]

plt.bar(courses, students, color=['blue','green','red','purple'])

plt.title("Course Enrollment")
plt.xlabel("Courses")
plt.ylabel("Students")

plt.show()

# Output:
# Bar chart showing students per course.
```

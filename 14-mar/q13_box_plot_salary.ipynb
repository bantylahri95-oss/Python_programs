```python
# Question: Box plot for salary distribution.

import matplotlib.pyplot as plt

dept1 = [30000,35000,40000,45000,50000]
dept2 = [25000,30000,35000,40000,42000]
dept3 = [20000,22000,25000,27000,30000]

plt.boxplot([dept1,dept2,dept3], labels=["HR","IT","Finance"])

plt.title("Salary Distribution")
plt.xlabel("Departments")
plt.ylabel("Salary")

plt.show()

# Output:
# Box plot showing salary spread.
```

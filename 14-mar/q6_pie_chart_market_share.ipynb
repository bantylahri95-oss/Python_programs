```python
# Question: Pie Chart for smartphone market share.

import matplotlib.pyplot as plt

brands = ["Apple","Samsung","Xiaomi","Oppo","Others"]
share = [30,25,20,15,10]

plt.pie(share, labels=brands, autopct='%1.1f%%')

plt.title("Market Share")

plt.show()

# Output:
# Pie chart showing percentage share.
```

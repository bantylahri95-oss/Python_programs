```python
# Question: Create subplots with Line, Bar, Pie, Scatter charts.

import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,2)

# Line
axs[0,0].plot([1,2,3],[10,20,30])
axs[0,0].set_title("Line")

# Bar
axs[0,1].bar(["A","B","C"],[5,7,3])
axs[0,1].set_title("Bar")

# Pie
axs[1,0].pie([40,30,30], labels=["X","Y","Z"], autopct='%1.1f%%')
axs[1,0].set_title("Pie")

# Scatter
axs[1,1].scatter([1,2,3],[4,5,6])
axs[1,1].set_title("Scatter")

plt.tight_layout()
plt.show()

# Output:
# Figure with 4 subplots (Line, Bar, Pie, Scatter).
```

```python
# Question: Histogram of marks.

import matplotlib.pyplot as plt
import numpy as np

marks = np.random.randint(40,100,100)

plt.hist(marks, bins=10, color='skyblue')

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()

# Output:
# Histogram showing distribution of marks.
```

```python
# Question: Bubble chart for population vs area vs GDP.

import matplotlib.pyplot as plt

area = [1484,603,426,206,709]
population = [19,20,11,14,12]
gdp = [300,350,200,220,280]

plt.scatter(area, population, s=[g*5 for g in gdp], alpha=0.5)

plt.title("Population vs Area (Bubble = GDP)")
plt.xlabel("Area")
plt.ylabel("Population")

plt.show()

# Output:
# Bubble chart with size representing GDP.
```

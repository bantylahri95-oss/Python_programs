```python
# Question: Plot a Multiple Line Chart showing sales of Product A, B, C.

import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]
A = [100,150,200,250,300,350]
B = [80,120,160,200,240,280]
C = [90,140,180,220,260,300]

plt.plot(months, A, label="Product A", marker='o')
plt.plot(months, B, label="Product B", marker='s')
plt.plot(months, C, label="Product C", marker='^')

plt.title("Product Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()

plt.show()

# Output:
# Multiple line chart comparing three products.
```

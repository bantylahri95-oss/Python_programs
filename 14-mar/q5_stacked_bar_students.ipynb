```python
# Question: Stacked Bar Chart for male and female students.

import matplotlib.pyplot as plt

dept = ["CS","IT","ECE","ME"]
male = [50,40,45,35]
female = [30,35,25,20]

plt.bar(dept, male, label="Male")
plt.bar(dept, female, bottom=male, label="Female")

plt.title("Students Distribution")
plt.xlabel("Departments")
plt.ylabel("Students")
plt.legend()

plt.show()

# Output:
# Stacked bars showing male and female counts.
```

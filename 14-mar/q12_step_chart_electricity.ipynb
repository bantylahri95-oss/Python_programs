```python
# Question: Step chart for electricity usage.

import matplotlib.pyplot as plt

time = ["6AM","9AM","12PM","3PM","6PM","9PM"]
usage = [2,5,7,6,8,4]

plt.step(time, usage)

plt.title("Electricity Consumption")
plt.xlabel("Time")
plt.ylabel("Units")

plt.show()

# Output:
# Step chart showing consumption pattern.
```

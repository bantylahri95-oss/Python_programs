```python
# Question: Scatter plot of study hours vs marks.

import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]
marks = [35,40,50,60,65,70,80,90]

plt.scatter(hours, marks, color='red')

plt.title("Study vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()

# Output:
# Scatter plot showing relationship.
```
